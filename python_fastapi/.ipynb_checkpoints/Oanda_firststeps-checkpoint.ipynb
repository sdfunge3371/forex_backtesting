```python
df = pd.DataFrame()
all_data = []  # 用來儲存每個 pair 的 DataFrame
combined_df = pd.DataFrame()
#"2025-04-30"
for pair in popular_list:
    print(pair)
    df = api.get_history(pair, "2021-01-01", "2025-04-30", "H6", "M")["c"]   
    print(type(df)) 
    df = df.to_frame(name=pair.replace("_", ""))  # 例如 EURUSD
    df.index = pd.to_datetime(df.index)

    if combined_df.empty:
        combined_df = df
    else:
        combined_df = combined_df.join(df, how="outer")

# 最後儲存
combined_df = combined_df.reset_index()
combined_df = combined_df.rename(columns={"index": "time"})
print(combined_df.head())
```
